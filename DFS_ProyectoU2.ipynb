```python
class Stack:
    #Constructor que se ejecuta al crear una nueva pila
    def __init__(self):
      #lista vacia para guardar elementos
        self.items = []
    #Para agreagar un elemento a la parte superior de la pila
    def push(self, x):
    #append() para añadir al elemento x al final de la lista
        self.items.append(x)

    #Para sacar y devolver el elemento que está hasta arriba de la pila
    def pop(self):
        if self.is_empty():
            raise IndexError("Pop de una pila vacía")
        return self.items.pop()
    #Para ver al elemento que está hasta  arriba de la pila sin sacarlo de la pila
    def peek(self):
        return self.items[-1] if not self.is_empty() else None

    #Saber si la pila está vacia
    def is_empty(self):
        return len(self.items) == 0

    #Para saber cuantos elementos hay en la pila
    def size(self):
        return len(self.items)
```

```python
class Graph:
    def __init__(self, directed: bool = False):

        self.directed = directed
        #Crea un diccionario vacio (lista de adyacencia)
        self.adj = {}  # dict: vertex -> set(neighbors)

  #Para añadir un nuevo vertice
    def add_vertex(self, v):
        if v not in self.adj:
            self.adj[v] = set()
  #Para crear una conexion entre un punto u y un punto v
    def add_edge(self, u, v):
        self.add_vertex(u)
        self.add_vertex(v)

        self.adj[u].add(v)
        if not self.directed:
            self.adj[v].add(u)
  #Para pedir la lista de vecinos de un método especifico
    def neighbors(self, v):
        return self.adj.get(v, set())

  #Determina como se verá el grafo cuando se imprima
    def __str__(self):
        lines = []
        for v in self.adj:
            nbrs = ", ".join(map(str, sorted(self.adj[v])))
            lines.append(f"{v}: [{nbrs}]")
        return "\n".join(lines)

    # DFS Iterativo
    def dfs_iterative(self, start):
            visited = set()
            visit_order = []
            s = Stack()
            s.push(start)

            while not s.is_empty():
              current = s.pop()

              if current not in visited:
                visited.add(current)
                visit_order.append(current)

                #Ordenamos al reves para que en la pila quede arriba el primer vecino en orden alfabetico.
                neighbors = sorted(self.neighbors(current), reverse=True)

                for neighbor in neighbors:
                  if neighbor not in visited:
                    s.push(neighbor)
            return visit_order

    #Componentes conexas
    def connected_components(self):
      global_visited = set()
      components = []
      for vertex in self.adj:
          if vertex not in global_visited:
            #Si encontramos un nodo nuevo, exploramos toda su "isla" conectada
            island = self.dfs_iterative(vertex)
            #Guardamos la isla entera en la lista maestra
            components.append(island)
            #Marcamos todos los nodos de la isla como visitados
            global_visited.update(island)
      return components

    #Deteccion de ciclos
    def has_cycle_undirected(self):
        global_visited = set()

        for start_vertex in self.adj:
            if start_vertex in global_visited:
                continue
            s = Stack()
            #Guardamos una tupla (nodo actual y nodo padre)
            s.push((start_vertex, None))

            while not s.is_empty():
                current, parent = s.pop()

                if current not in global_visited:
                    global_visited.add(current)

                    for neighbor in self.neighbors(current):
                        if neighbor not in global_visited:
                            s.push((neighbor, current))

                        #Si el vecino ya está visitado y no es donde venimos a buscar entonces hay un ciclo
                        elif neighbor != parent:
                            return True
        return False
```

```python
# PRUEBA 1
# DFS en grafo sin ciclo
g = Graph(directed=False)
edges = [("A", "B"), ("A", "C"), ("C", "D")]
for u, v in edges:
    g.add_edge(u, v)
print("DFS sin ciclo")
print("GRAFO 1")
print(g)
```

```python
print("Orden de visita")
print(g.dfs_iterative("A"))
```

```python
print("¿Hay un ciclo cerrado?")
print(g.has_cycle_undirected())
```

```python
#PRUEBA 2
#Detección de ciclo
g2 = Graph(directed=False)
edges2 = [("A", "B"), ("B", "C"), ("C", "A")]
for u, v in edges2:
    g2.add_edge(u, v)

print("GRAFO 2")
print(g2)
```

```python
print("¿Hay un ciclo cerrado?")
print(g2.has_cycle_undirected())
```

```python
#Componentes conexas
# Conexiones
# Isla 1: A y B
# Isla 2: C, D y E
# Isla 3: F (completamente solo)

g3 = Graph(directed=False)
edges3 = [("A", "B"), ("C", "D"), ("D", "E")]
for u, v in edges3:
    g3.add_edge(u, v)
g3.add_vertex("F")

print("GRAFO 3")
print(g3)
```

```python
print(g3.connected_components())
```
